# Interesting visual examples with Seaborn + Pandas

## Seaborn

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%pylab inline
```

Create dataset

```python
num = pd.DataFrame(np.random.normal(0,1,size=(100,2)))
ss = pd.DataFrame(np.random.normal(0,1,size=(100,1)))
dat = np.hstack((ss,ss))
datt = np.vstack((dat,num))
```

Linear regression graph. Here you can see that regression is very sensitive to noise.

```python
sns.regplot(datt[:,0],datt[:,1])
plt.show()
```

Density plot and smooth density plot

```python
sns.distplot(ss);
sns.kdeplot(datt[:,0]);
```

Two-dimensional joint and separate distribution

```python
sns.kdeplot(datt);
sns.jointplot(datt[:,0],datt[:,1], kind='kde');
```

Hex plot with same data. In the upper right corner you can see Spearman (Pearson) coefficient

```python
sns.jointplot(datt[:,0],datt[:,1], kind='hex');
#kind = 'hex','reg','kde'
#stat_func = spearmanr
```

## Pandas

Now we create data with 3 features. Try to picture simple and cumulative graphics. You can do this with .bar method for vertical and .barh for horizontal picture. Bar plot can show you difference between some groups in your data.

```python
aa = pd.DataFrame(np.random.normal(0,1,size=(10,3)))
aa.plot.bar(figsize=(10,6))
plt.show()
print('Cumulative graph')
aa.plot.bar(figsize=(10,6),stacked=True)
plt.show()
print('Cumulative horizontal graph')
aa.plot.barh(figsize=(10,6),stacked=True)
plt.show()
```

It's boxplot time. Remember, that boxplot don't show distribution structure (polymodal, asymmetric or not, skew or not and so). But boxplot helps to compare different data distributions. It's easy to compare min, max, mean, quantiles with boxplot. Sometimes you can see difference with boxplot without special statistic hypothesis. 

```python
aa.plot.box(figsize=(10,6));
```

Area plot to compare data.

```python
aa = pd.DataFrame(np.random.randint(1,5,size=(50,5)))
aa.plot.area(figsize=(10,6));
```

Pie plot. But keep in mind, that information through pie chart is perceived worse by human than metric plot.

```python
aa = pd.DataFrame(np.random.randint(1,5,size=(10,1)))
aa.plot.pie(figsize=(8,8), subplots=True);
```
